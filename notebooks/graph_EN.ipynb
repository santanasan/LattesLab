# CREATE GRAPH FROM LATTES CV COLLECTIONS IN A FOLDER

The code in this notebook extracts from Lattes CV files the researchers that were co-authors with the Lattes CV owner in scientific publications.

Explanations for the code will follow each cell. The cell below is reserved for definitions and load of Python packages.

The variable "folder" contains the name of the folder where the Lattes CV collection is found.

The code in the following cell reads each Lattes CV in the folder and creates a network of nodes. Each node is a researcher and each edge between nodes means that the researchers of those two nodes have published together.

It is expected that the researchers whose Lattes CVs are found in the folder will contain higher number of edges.

The function classifies the owners of the Lattes CVs found in the folder as part of a group - with the label 'internal'. Other researchers found in the Lattes CVs are classified as outsiders - with the label 'external'.

The network graph is plotted in two different arrangements, the first one called the spring layout. The blue nodes are the researchers in the internal group, and the red nodes are the ones in the external group. The second one is called shell layout. It is possible to arrange the internal group inside and the external group outside, to have a notion of the edges between the groups.

The function also saves the figure and the correlation between the integers associated with each node and the respective name of the researcher.

```python
[graph, graphframe] = ll.get_graph_from_folder(folder)
```
